```python
#checking the no. of types available for each category

def count_types(folder_dir):
    for subfolder in os.listdir(folder_dir):
        count = 0
        for sub in os.listdir(os.path.join(folder_dir, subfolder)):
            count += 1
        print(subfolder, "has", count, "different models.")
```
